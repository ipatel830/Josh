```python
from dataset import process_data
from transformers import Wav2Vec2CTCTokenizer
```

```python
processed = process_data(root_folder='LibriSpeech/train-460',save_file='train_tokenized.pt',train=True)
```

```python
tokenizer = Wav2Vec2CTCTokenizer.from_pretrained("./wav2vec2_tokenizer")
data_len,missing_audio  = processed.tokenize(tokenizer)
```

```python
process_test = process_data(root_folder='LibriSpeech/test-clean',save_file='',train=False)
data_len,missing_audio = process_test.tokenize(tokenizer)
```

```python
print(f"Length of data: {data_len}")
print(f"Missing audio: {missing_audio}")
```

```python
with open('output.txt','w') as f:
    f.write(f'Length of dataset: {data_len}')
    f.write(f"Missing audio: {missing_audio}")
    f.close()
```
